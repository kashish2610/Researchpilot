```python
from langchain_community.tools import DuckDuckGoSearchRun
search_tool = DuckDuckGoSearchRun()
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/NextG-GPT_Leveraging_GenAI_for_Advancing_Wireless_.pdf")
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=150
)

chunks = splitter.split_documents(docs)
```

```python
import langchain
print(langchain.__version__)
print(langchain.__file__)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(chunks, embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from google.colab import userdata

hf_token = userdata.get("HF_TOKEN")

endpoint = HuggingFaceEndpoint(
    repo_id="meta-llama/Meta-Llama-3-8B-Instruct",
    task="text-generation",
    max_new_tokens=800,
    temperature=0.2,
    huggingfacehub_api_token=hf_token
)

llm = ChatHuggingFace(llm=endpoint)
```

```python
from langchain_core.prompts import PromptTemplate

decision_prompt = PromptTemplate(
    input_variables=["question"],
    template="""
You are an expert research assistant.

Decide whether answering the following question
requires up-to-date information from the web
or can be answered using existing research papers.

Question:
{question}

Answer ONLY with one word:
YES or NO
"""
)
decision_agent = decision_prompt | llm

```

```python
def needs_web_search(question):
    decision = decision_agent.invoke({"question": question}).content.strip().upper()
    return decision == "YES"
```

```python
def web_search_context(query, max_results=5):
    results = search_tool.invoke(query)
    return results
```

```python
def needs_web_search(question):
    decision = decision_agent.invoke({"question": question}).content.strip().upper()
    return decision == "YES"
```

```python
web_reader_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a technical research assistant.

Using ONLY the web search context below,
extract the most relevant and factual information
related to the question.

Context:
{context}

Question:
{question}

Return concise bullet points.
"""
)

web_reader_agent = web_reader_prompt | llm
```

## Query Decomposition Agent

```python
from langchain_core.prompts import PromptTemplate
decompose_query= PromptTemplate(
    input_variables=["question"],
    template=""""
Generate 3 different search queries
that capture different technical interpretations
of the following research question.

Question:
{question}
"""

)
query_agent= decompose_query | llm
```

## Retriever Agent

```python
def retrive_docs(sub_query):
    docs = retriever.invoke(sub_query)
    """print("Sub Query:", sub_query)
    print("Number of docs retrieved:", len(docs))
    print("----- SAMPLE DOC -----")

    if len(docs) > 0:
        print(docs[0].page_content[:500])  # first 500 chars
    else:
        print("No documents retrieved.")"""
    return docs
```

## Reader Agent (Evidence Extractor)

```python
reader_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""You are a research paper analyst.

Context:
{context}

Question:
{question}

Extract factual points with citations.
"""
)

reader_agent = reader_prompt | llm
```

## Critic Agent (Validation)

```python
critic_ptompt= PromptTemplate(
    input_variables=["context", "question", "answer"],
    template="""you are a critical review agent.

Context:
{context}

Question:
{question}

Answer:
{answer}
Return corrected version.
"""
)
critic_agent = critic_ptompt | llm
```

## Synthesizer Agent (Final Answer)

```python
synth_prompt=PromptTemplate(
    input_variables=["answer"],
    template="""Combine the following expert answers into one coherent response.
Ensure clarity, citations, and structured explanation.Provide the answer in 1000 words with context completion.

Answers:
{answers}
"""
)
synthesizer_agent = synth_prompt | llm
```

## Multi-Agent Orchestration

```python
def multi_agent_rag_with_web(question):
    # Step 1: Decompose
    sub_queries = query_agent.invoke({"question": question}).content.split("\n")

    all_answers = []

    for q in sub_queries:
        current_context = ""
        # Step 2: Decide web vs papers
        if needs_web_search(q):
            web_context = web_search_context(q)
            current_context = web_context
            answer = web_reader_agent.invoke({
                "context": web_context,
                "question": q
            })
        else:
            docs = retrive_docs(q)
            context = "\n".join([d.page_content for d in docs])
            current_context = context
            answer = reader_agent.invoke({
                "context": context,
                "question": q
            })

        reviewed = critic_agent.invoke({
            "context": current_context,
            "question": q,
            "answer": answer.content
        })

        # Re-add truncation for reviewed.content to a maximum of 1500 characters
        truncated_content = reviewed.content[:1500] if len(reviewed.content) > 1500 else reviewed.content
        all_answers.append(truncated_content)

    print(f"Length of combined answers before synthesizing: {len('\n'.join(all_answers))}")
    final = synthesizer_agent.invoke({
        "answers": "\n".join(all_answers)
    })

    return final.content
```

```python
question = "what work GEN AI is doing wireless communcation provide me its purpose and role of this NextG gpt and how RAG is working ?"
print(multi_agent_rag_with_web(question))
```
